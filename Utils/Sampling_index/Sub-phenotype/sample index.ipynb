```python
import pickle
import numpy as np
from sklearn.model_selection import train_test_split
import os
import pandas as pd
```

```python
path_dataset = '.../graph_output_bmi_1&2_sub/total_enc_encode.pickle'
file_dataset = open(path_dataset, 'rb')
df_features = pickle.load(file_dataset)
```

```python
df_features = df_features[['PATID', 'BMI_DIFF']]
print(df_features['BMI_DIFF'].value_counts())
df_features.index = range(len(df_features))
PATID_list = df_features['PATID']
# get unique PATID and splite the based on PATID
unique_PATID_list = PATID_list.unique()
train_ratio = 0.7
validation_ratio = 0.1
test_ratio = 0.2

# First split to separate out the training set
train_pat, temp_set = train_test_split(unique_PATID_list, test_size=(1 - train_ratio), random_state=42)

# Second split to divide the temp_set into validation and test sets
val_test_ratio = test_ratio / (validation_ratio + test_ratio)
val_pat, test_pat = train_test_split(temp_set, test_size=val_test_ratio, random_state=42)

train_df = df_features[df_features['PATID'].isin(train_pat)]
val_df = df_features[df_features['PATID'].isin(val_pat)]
test_df = df_features[df_features['PATID'].isin(test_pat)]
```

```python
print('#patient in train: ', len(train_df['PATID'].unique()), '# encounter: ',len(train_df))
print('#patient in val: ', len(val_df['PATID'].unique()), '# encounter: ',len(val_df))
print('#patient in test: ', len(test_df['PATID'].unique()), '# encounter: ',len(test_df))
```

```python
label = train_df['BMI_DIFF'].astype(int)
```

```python
subset_index, _, sub_label, _= train_test_split(train_df.index, label, train_size=0.1, stratify=label , random_state=42)
print(subset_index)
sub_label.value_counts()
```

```python
import pandas as pd
index_df = pd.DataFrame(train_df.index, columns=["index"])
index_df.to_csv("train_indices.csv", index=False)

index_df = pd.DataFrame(val_df.index, columns=["index"])
index_df.to_csv("val_indices.csv", index=False)

index_df = pd.DataFrame(test_df.index, columns=["index"])
index_df.to_csv("test_indices.csv", index=False)
```
